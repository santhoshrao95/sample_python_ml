```python
def generating_samples(input_data, target_data):

    '''In this function, we will write code for generating 30 samples '''
    # you can use random.choice to generate random indices without replacement
    # Please have a look at this link https://docs.scipy.org/doc/numpy-1.16.0/reference/generated/numpy.random.choice.html for more details
    # Please follow above pseudo code for generating samples 
    
    selecting_rows = np.random.choice(x.shape[0],size=303,replace=False)
    #print(selecting_rows)
    replacing_rows = np.random.choice(selecting_rows,size=203,replace=True)

    size = random.sample(range(3,x.shape[1]),1)[0]
    selecting_cols = np.random.choice(x.shape[1],size=size,replace=False)

    sample_data = input_data[selecting_rows][:,selecting_cols]

    target_of_sample_data = target_data[selecting_rows]

    #replicating data
    replicated_sample_data = input_data[replacing_rows][:,selecting_cols]
    target_of_replicated_sample_data = target_data[replacing_rows]

    #print(sample_data.shape)
    #print(target_of_sample_data.shape)
    #print(replicated_sample_data.shape)
    #print(target_of_replicated_sample_data.shape)

    #concatinating data
    sampled_input_data = np.vstack((sample_data,replicated_sample_data))
    sampled_target_data = np.vstack((target_of_sample_data.reshape(-1,1),target_of_replicated_sample_data.reshape(-1,1)))

    return list(sampled_input_data) , list(sampled_target_data), list(selecting_rows),list(selecting_cols)
    #note please return as lists

```

```python
def build_models(list_input_data,list_output_data):
  list_of_all_models = []
  for i in range(0,30):
    model = DecisionTreeRegressor(max_depth=None)
    list_of_all_models.append(model.fit(list_input_data[i],list_output_data[i]))
  return list_of_all_models
```

<font color='orange'><b>Flowchart for calculating MSE </b></font>

![alt text](https://i.imgur.com/sPEE618.png)

After getting predicted_y for each data point, we can use sklearns mean_squared_error to calculate the MSE between predicted_y and actual_y.

*  <font color='blue'><b> Write code for calculating MSE</b></font>

```python
def calculate_mse(list_input_data,list_of_all_models):
  y_pred = []
  for i in range(len(x)):
    y_k = []
    for k in range(len(list_of_all_models)):
      y_k.append(list_of_all_models[k].predict(list_input_data[k][i].reshape(1,-1)))
    y_pred.append(np.mean(y_k))  
  mse = mean_squared_error(y,y_pred)
  return mse
```

<font color='blue'><b>Step - 3 </b></font>

<font color='orange'><b>Flowchart for calculating OOB score</b></font>

![alt text](https://i.imgur.com/95S5Mtm.png)

Now calculate the $OOB Score =  \frac{1}{506}\sum_{i=1}^{506}(y^{i} - y^{i}_{pred})^{2}$.

*  <font color='blue'><b> Write code for calculating OOB score </b></font>

```python
def calculate_oobscore(list_input_data,list_of_all_models,list_selected_row):
  oob_pred = []
  for i in range(len(x)):
    y_k=[]
    for k in range(len(list_of_all_models)):
      if i not in list_selected_row[k]:
        y_k.append(list_of_all_models[k].predict(list_input_data[k][i].reshape(1,-1)))
    oob_pred.append(np.mean(y_k))  
  oobscore = mean_squared_error(y,oob_pred)
  return oobscore
```

```python
mse = []
oobs = []
for j in tqdm(range(35)):
  # Use generating_samples function to create 30 samples 
  # store these created samples in a list
  list_input_data =[]
  list_output_data =[]
  list_selected_row= []
  list_selected_columns=[]

  for i in range(0,30):
    a,b,c,d = generating_samples(x, y)
    list_input_data.append(a)
    list_output_data.append(b)
    list_selected_row.append(c)
    list_selected_columns.append(d)

  list_of_all_models = build_models(list_input_data,list_output_data)
  mse.append(calculate_mse(list_input_data,list_of_all_models))
  oobs.append(calculate_oobscore(list_input_data,list_of_all_models,list_selected_row))

```

```python
from prettytable import PrettyTable
def CI(data):
  ci = PrettyTable()
  ci = PrettyTable(["Sample Size", "Sample mean","Left C.I","Right C.I"])
  sample=data[random.sample(range(0, data.shape[0]), 30)]
  sample_mean = sample.mean()
  sample_std =  sample.std()
  sample_size = len(sample)
  # here we are using sample standard deviation instead of population standard deviation
  left_limit  = np.round(sample_mean - 2*(sample_std/np.sqrt(sample_size)), 3)
  right_limit = np.round(sample_mean + 2*(sample_std/np.sqrt(sample_size)), 3)
  row = []
  #row.append(i+1)
  row.append(sample_size)
  row.append(sample_mean)
  #row.append(population_std)
  row.append(left_limit)
  row.append(right_limit)
  #row.append(male_population_mean)
  #row.append((male_population_mean <= right_limit) and (male_population_mean >= left_limit))
  ci.add_row(row)
  print(ci)
```
